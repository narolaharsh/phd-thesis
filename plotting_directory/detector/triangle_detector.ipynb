```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science'])
import matplotlib.patches as patches
import matplotlib.transforms as transforms
import matplotlib as mpl

```

```python

fig, ax = plt.subplots(1, 1, figsize = (4, 4))

### make lines ##########

r = 1
trim = 0.07
zorder_line = 0
#### detector 1 ###
r1 = 1-0.07
ax.plot([0, r1], [0, 0], color = 'lightblue', zorder = zorder_line)
ax.plot([0, r1*np.cos(np.deg2rad(60))], [0, r1*np.sin(np.deg2rad(60))], color = 'lightblue', zorder = zorder_line)

detector_1 = {'origin': [0, 0], 'arm_left': [r1*np.cos(np.deg2rad(60)),r1*np.sin(np.deg2rad(60))], 'arm_right': [r1, 0] }


#### detector 2 ###
origin_x = r + (trim*np.cos(np.deg2rad(60)))
origin_y = 0. + (trim*np.sin(np.deg2rad(60)))
ax.plot([origin_x, origin_x-r1], [origin_y, origin_y-0], color = 'lightgreen', zorder = zorder_line)

ax.plot([origin_x, origin_x - r1*np.cos(np.deg2rad(60))], [origin_y, origin_y+r1*np.sin(np.deg2rad(60))], color = 'lightgreen', zorder = zorder_line)

detector_2 = {'origin': [origin_x, origin_y], 'arm_left': [origin_x-r1, origin_y], 'arm_right': [origin_x - r1*np.cos(np.deg2rad(60)), origin_y+r1*np.sin(np.deg2rad(60))]}

#### detector 3 ###
origin_x = r/2  - trim*np.cos(np.deg2rad(60))
origin_y = r*np.sin(np.deg2rad(60)) + (trim*np.sin(np.deg2rad(60)))

ax.plot([origin_x, origin_x-r1*np.cos(np.deg2rad(60))], [origin_y, origin_y-r1*np.sin(np.deg2rad(60))], color = 'salmon', zorder = zorder_line)
ax.plot([origin_x, origin_x+r1*np.cos(np.deg2rad(60))], [origin_y, origin_y-r1*np.sin(np.deg2rad(60))], color = 'salmon', zorder = zorder_line)
detector_3 = {'origin': [origin_x, origin_y], 'arm_left': [origin_x-r1*np.cos(np.deg2rad(60)), origin_y-r1*np.sin(np.deg2rad(60))], 'arm_right': [origin_x+r1*np.cos(np.deg2rad(60)), origin_y-r1*np.sin(np.deg2rad(60))]}
```

```python

########## Make mirrors #######################

loc = [detector_3['arm_left'][0], detector_3['arm_left'][1]]
mirror_color = 'oldlace'
mirror_size = 40
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(60)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)

ax.scatter(loc[0]+0.8*np.cos(np.deg2rad(60)), loc[1]+0.8*np.sin(np.deg2rad(60)), marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0]+0.8*np.cos(np.deg2rad(60)), loc[1]+0.8*np.sin(np.deg2rad(60)), s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)


loc = [detector_3['arm_right'][0], detector_3['arm_right'][1]]
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(120)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
```

```python

ax.scatter(loc[0]-0.8*np.cos(np.deg2rad(60)), loc[1]+0.8*np.sin(np.deg2rad(60)), marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0]-0.8*np.cos(np.deg2rad(60)), loc[1]+0.8*np.sin(np.deg2rad(60)), s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
########################### mirrors on detector 2 ########################
loc = [detector_2['arm_left'][0], detector_2['arm_left'][1]]
mirror_color = 'oldlace'
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(0)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.scatter(loc[0]+0.8, loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)

ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
ax.text(loc[0]+0.8, loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)

loc = [detector_2['arm_right'][0], detector_2['arm_right'][1]]
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(120)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.scatter(loc[0]+0.8*np.cos(np.deg2rad(60)), loc[1]-0.8*np.sin(np.deg2rad(60)), marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
ax.text(loc[0]+0.8*np.cos(np.deg2rad(60)), loc[1]-0.8*np.sin(np.deg2rad(60)), s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)



########################### mirrors on detector 1 ########################
loc = [detector_1['arm_left'][0], detector_1['arm_left'][1]]
mirror_color = 'oldlace'
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(60)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.scatter(loc[0]-(0.8*np.cos(np.deg2rad(60))), loc[1]-(0.8*np.sin(np.deg2rad(60))), marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
ax.text(loc[0]-(0.8*np.cos(np.deg2rad(60))), loc[1]-(0.8*np.sin(np.deg2rad(60))), s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)

loc = [detector_1['arm_right'][0], detector_1['arm_right'][1]]
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(0)
ax.scatter(loc[0], loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.scatter(loc[0]-0.8, loc[1], marker = s, s = mirror_size, color = mirror_color, edgecolor = 'black', lw = 0.8)
ax.text(loc[0], loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)
ax.text(loc[0]-0.8, loc[1], s = 'M', ha = 'center', va = 'center', size = 4, rotation = 60)


########## mirror section complete #######################
```

```python


############# beam splitter ##############
splitter_size = 40
splitter_facecolor = 'mintcream'
splitter_edgecolor = 'black'
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(45)
ax.scatter(detector_1['origin'][0], detector_1['origin'][1], marker = s, s =  splitter_size, 
           facecolor = splitter_facecolor,
           edgecolor = splitter_edgecolor)
ax.text(detector_1['origin'][0], detector_1['origin'][1], s = 'B', ha = 'center', va = 'center', size = 4)


s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(45)
ax.scatter(detector_2['origin'][0], detector_2['origin'][1], marker = s, s =  splitter_size,
            facecolor = splitter_facecolor,
           edgecolor = splitter_edgecolor)
ax.text(detector_2['origin'][0], detector_2['origin'][1], s = 'B', ha = 'center', va = 'center', size = 4)



s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(0)
ax.scatter(detector_3['origin'][0], detector_3['origin'][1], marker = s, s =  splitter_size,
            facecolor = splitter_facecolor,
           edgecolor = splitter_edgecolor)
ax.text(detector_3['origin'][0], detector_3['origin'][1], s = 'B', ha = 'center', va = 'center', size = 4)

```

```python



############# Lasers sources ##############
lasercolor = 'lightblue'

s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(60)
source_size = 70
laser_1_loc = [detector_1['origin'][0]-(0.2*np.cos(np.deg2rad(60))), detector_1['origin'][1]-(0.2*np.sin(np.deg2rad(60)))]
ax.scatter(laser_1_loc[0], laser_1_loc[1], marker = s, color = lasercolor, s = source_size, edgecolor='black')
ax.text(laser_1_loc[0], laser_1_loc[1], s = 'LS', ha = 'center', va = 'center', size = 4)


transpose_loc = np.array([detector_1['origin'], laser_1_loc]).T

ax.plot(transpose_loc[0], transpose_loc[1], color = lasercolor, zorder = 0)

############# Lasers sources ##############
lasercolor = 'lightgreen'
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(30)
laser_2_loc = [detector_2['origin'][0]+(0.2*np.cos(np.deg2rad(60))), detector_2['origin'][1]-(0.2*np.sin(np.deg2rad(60)))]
ax.scatter(laser_2_loc[0], laser_2_loc[1], marker = s, color = lasercolor, s = source_size, edgecolor='black')
transpose_loc = np.array([detector_2['origin'], laser_2_loc]).T
ax.text(laser_2_loc[0], laser_2_loc[1], s = 'LS', ha = 'center', va = 'center', size = 4)


ax.plot(transpose_loc[0], transpose_loc[1], color = lasercolor, zorder = 0)

############# Lasers sources ##############

lasercolor = 'salmon'
s = mpl.markers.MarkerStyle(marker='s')
s._transform = s.get_transform().rotate_deg(30)
laser_3_loc = [detector_3['origin'][0] - 0.2*np.cos(np.deg2rad(60)), detector_3['origin'][1]+0.2*np.sin(np.deg2rad(60))]

ax.scatter(laser_3_loc[0], laser_3_loc[1], marker = s, color = lasercolor, edgecolor='black', s = source_size)
ax.text(laser_3_loc[0], laser_3_loc[1], s = 'LS', ha = 'center', va = 'center', size = 4)

transpose_loc = np.array([detector_3['origin'], laser_3_loc]).T

ax.plot(transpose_loc[0], transpose_loc[1], color = lasercolor, zorder = 0)


```
