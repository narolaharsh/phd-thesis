```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science'])
plt.rcParams['axes.labelsize'] = 11.
plt.rc('xtick',labelsize=11)
plt.rc('ytick',labelsize=11)
```

```python
red = (0.80, 0.36, 0.36)
green = (0.62, 0.79, 0.62)
blue = (0.27, 0.51, 0.71)



lw = 1.5

bins = np.arange(3, 9, 1)
values = np.array([3, 4, 4, 5, 5, 5, 6, 6, 7])

fine_bins = np.arange(0, 10, 0.01)
sigma2 = 2.5
gaussian = 3 * np.exp(-1* np.power(fine_bins - 5, 2)/sigma2)

print(bins)

blue = '#A3BFD9'
fig, ax = plt.subplots(2, 1, sharex = True, figsize = (4, 6), gridspec_kw={'hspace': .05})
ax[0].hist(values, bins = bins, histtype = 'bar', align = 'left', edgecolor = 'white', color = blue, alpha = .5)
for ii in range(len(bins)):
    x = bins[ii]
    y = 3 - abs(x - 5)
    ax[0].scatter(np.full(shape=y, fill_value=x), np.linspace(0.5, y-0.5, y), color = 'black')


ax[0].plot(fine_bins, gaussian, color = (0.27, 0.51, 0.8), lw = 1.5, ls = '--')

seed = 7
np.random.seed(seed)
y = np.random.uniform(0, 100, size = len(values))

failed_y = np.random.uniform(0, 100, size = 3)

failed_values = np.random.choice(np.arange(bins[0], bins[-1], 1), size = (len(failed_y)), replace=True)
failed_values = np.array([float(x) for x in failed_values])
failed_values += np.random.uniform(-0.2, 0.2, size = len(failed_values))

accepted_values = values + np.random.uniform(-0.3, 0.3, size = len(values))
ax[1].scatter(accepted_values, y, color = 'black')
ax[1].scatter(failed_values, failed_y, marker = 'o', facecolor = 'none', edgecolor = 'black')
#ax[1].set_ylabel('Iterations')

for b in bins:
    ax[1].axvline(x = b-0.5, color = 'grey', alpha = 0.2)


print(accepted_values)
print(y)


sorting = np.argsort(accepted_values)

accepted_val_sort = accepted_values[sorting]
y_sort = y[sorting]

for ii in range(len(accepted_val_sort)-1):

    ax[1].annotate("", xytext = (accepted_val_sort[ii], y_sort[ii]), xy = (accepted_val_sort[ii+1], y_sort[ii+1]), arrowprops=dict(arrowstyle="->", color = green, lw = lw))


failed_y_sorted = failed_y[np.argsort(failed_values)]
failed_values_sorted = failed_values[np.argsort(failed_values)]



ax[1].annotate("", xytext = (accepted_val_sort[1], y_sort[1]), xy = (failed_values_sorted[0], failed_y_sorted[0]), arrowprops=dict(arrowstyle="->", color = red, ls = 'dotted', lw = lw))
ax[1].annotate("", xytext = (accepted_val_sort[3], y_sort[3]), xy = (failed_values_sorted[1], failed_y_sorted[1]), arrowprops=dict(arrowstyle="->", color = red, ls = 'dotted', lw = lw))
ax[1].annotate("", xytext = (accepted_val_sort[6], y_sort[6]), xy = (failed_values_sorted[2], failed_y_sorted[2]), arrowprops=dict(arrowstyle="->", color = red, ls = 'dotted', lw = lw))


for axes in ax:
    axes.set_yticks([])
    axes.set_yticklabels([])

ax[1].set_xlabel("Parameter Space")
ax[1].set_xlim(2.2, 7.8)
ax[1].set_xticks([3, 4, 5, 6, 7])
ax[1].set_xticklabels(["$\\vec{\\theta}_1$", "$\\vec{\\theta}_2$", "$\\vec{\\theta}_3$", "$\\vec{\\theta}_4$", "$\\vec{\\theta}_5$"])
ax[0].set_ylabel('PDF')
ax[1].set_ylabel("Markov chain progression")
fig.savefig('./markov_chain.pdf')
fig.savefig('../../figures/markov_chain.pdf')



```
